## <a id='Data_Visualization'></a>Data Visualization ü§î
---
Data visualization is the graphical representation of information and data. Using visual elements such as charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

In the world of big data, data visualization tools and techniques are essential for analyzing vast amounts of information and making data-driven decisions.

- **For a simple overview, here are some popular plotting libraries:**

   - [Matplotlib](https://matplotlib.org/): Low level, provides a lot of freedom.   
     
   - [Pandas visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html): an easy-to-use interface, built on Matplotlib.  
    
   - [Seaborn](https://seaborn.pydata.org/): High-level interface, cool virtual styles.

## <a id='scatter'></a>Scatter Plot ü§î

- To generate a scatter plot in **Matplotlib** we can use the `scatter` method. We will also create a figure and axis with `plt.subplots` so we can give our title and labels.


- To create a scatter plot in **Pandas**, we can call `<dataset>.plot.scatter()` and pass two arguments, column name x as well as column name y. Optionally we can also pass an title to it.


- We can use the `.scatterplot` method to create a scatter plot in **Seaborn**, and just like in Pandas we need to pass the column names of x and y data, but now we also need to pass the data as an extra argument because we don't call the function on the data directly like we did in Pandas.

### Exercise 1 (Scatter Plot using Matplotlib) üëá
---

```python
####################################### Scatter Plot using Matplotlib #########################################

color=np.random.randint(0,10,150)
# scatter the SepalLengthCm against the SepalWidthCm
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'],
            s=150,   # Size
            c=color, # Sequence of n numbers to be mapped to colors using cmap
            cmap="Reds", # Instance colormap name. cmap is only used if c is an array of floats.
            edgecolor="black", # The edge color of the marker
            linewidth=1, # The linewidth of the marker edges
            alpha=1) # The alpha blending value

cbar=plt.colorbar()
# set a title and labels
plt.title('Iris Dataset')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
```

```python
colors = {'Iris-setosa':'b', 'Iris-versicolor':'tab:orange', 'Iris-virginica':'g'}
colors['Iris-setosa']
```

**Note**: We can give the graph more meaning by coloring each data point with its category.üëå

```python
# create color dictionary
colors = {'Iris-setosa':'b', 'Iris-versicolor':'tab:orange', 'Iris-virginica':'g'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i], iris['SepalWidthCm'][i],color=colors[iris['Species'][i]],
            s=150,   # Size
            edgecolor="black", # The edge color of the marker
            linewidth=1, # The linewidth of the marker edges
            alpha=1)   # The alpha blending value)
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
```

## <a id='line'></a>Line Plot üìà üìâ
---

- In **Matplotlib** we can create a line plot by calling the `plot` method. We can also plot multiple columns in a single graph, by looping through the columns we want and plotting each column on the same axis.


- To create a line plot in **Pandas**, we can call `<dataframe>.plot.line()`. While in Matplotlib we needed to loop over every column we wanted to plot, in Pandas we don't need to do that because it automatically plots all available numeric columns (at least if we don't define a specific column/s).


- To create a line plot in **Seaborn**, the `sns.lineplot` method can be used. The only argument required is the data, which in our case is the four numeric columns from the Iris dataset. We can also use the `sns.kdeplot` method which goes around the edges of the curves, so it's much cleaner if you have a lot of outliers in your dataset.

### Exercise 4 (Line Plot using Matplotlib) üëá
---

```python
####################################### Line Plot using Matplotlib #########################################
# get columns to plot
columns = iris.columns.drop(['Id','Species'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

## <a id='bar'></a>Bar Plot üìä

- A bar chart can be created in **Matplotlib** using the `bar` method. The bar chart does not automatically calculate the frequency of a category, so we will use the pandas value_counts function to do this. A bar chart is useful for categorical data that doesn't have a lot of different categories (less than 30) because otherwise it can get very cluttered.


- To plot a bar-chart in **Pandas** we can use the `plot.bar()` method, but before we can call this we need to get our data. For this we will first count the occurrences using the `value_count()` method and then sort the occurrences from smallest to largest using the `sort_index()` method.



- In **Seaborn**, a bar chart can be created using the `sns.countplot` method and pass data to it.

### Exercise 10 (Bar Plot using Matplotlib) üëá
---

```python
cp.head(1)
```

```python
####################################### Bar Plot using Matplotlib #########################################

# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = cp['gender'].value_counts()
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
for p in ax.patches:
    height = p.get_height()
    ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,s="{}%".format(height),ha='center')
ax.set_title('Gender Frequency')
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

## <a id='pie'></a>Pie Plot

- **Matplotlib** has `pie()` function in its `pyplot` module which create a pie chart representing the data in an array.



- To plot a pie-chart in **Pandas** we can use the `plot.pie()` method, but before we can call this we need to get our data. For this we will first group by the occurrences using the `groupby()` method and then select one column and get it's sum using `sum()` method.

---

- Exercise 13 (Pie Plot using Matplotlib)
- Exercise 14 (Pie Plot using Pandas visualization)
- Now Your Turn üí™üí™

---


### Exercise 13 (Pie Plot using Matplotlib)
---

```python
####################################### Pie Plot using Matplotlib #########################################

# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = cp.groupby("gender")["salary"].sum()
# get x and y data 
points = data.index 
frequency = data.values
# create bar chart 
ax.pie(frequency, labels= points, autopct="%.1f%%") 
# set title and labels 
ax.set_title('Gender pie chart')
ax.legend(points)
ax.set_ylabel('salary')
```

### Exercise 14 (Pie Plot using Pandas visualization)
---

```python
#################################### Pie Plot using Pandas visualization #######################################

#groupby the data by delivery type
data = cp.groupby("gender")["salary"].sum()
data.plot.pie(autopct="%.1f%%");
ax.set_title('Gender pie chart')
```

## <a id='stem'></a>Stem Plot

- To plot a stem plot in **Matplotlib**, we can use `stem()` method. It creates vertical lines from a baseline to the Y-coordinate and places a marker at the tip.

---

```python
####################################### Stem Plot using Matplotlib #########################################
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))


plt.stem(x, y, use_line_collection = True,linefmt='grey', markerfmt='*')
# set a title and labels
plt.title('Sin Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

markerline, stemlines, baseline = plt.stem(x, y, linefmt='grey', markerfmt='*', bottom=1.035)

markerline.set_markerfacecolor('red')
plt.title('Sin Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
